```python
from bs4 import BeautifulSoup
from urllib.request import urlopen

html = urlopen('https://news.google.com/topics/CAAqJQgKIh9DQkFTRVFvSUwyMHZNR3QwTlRFU0JYcG9MVlJYS0FBUAE?hl=zh-TW&gl=TW&ceid=TW%3Azh-Hant')
bsObj = BeautifulSoup(html.read(), 'html.parser')
```

```python
bsObj.article.a
```

```python
titleList = bsObj.findAll("a", {"class": "DY5T1d"})
for title in titleList:
    print(title.get_text())
```

```python
authorList = bsObj.findAll("a", {"class": "wEwyrc AVN2gc uQIVzc Sksgp"})
for author in authorList:
    print(author.get_text())
```

```python
timeList = bsObj.findAll("time", {"class": "WW6dff uQIVzc Sksgp"})
for time in timeList:
    print(time.get_text())
```

### 一次擷取整個新聞相關資訊

```python
articleList = bsObj.findAll("div", {"class": "NiLAwe y6IFtc R7GTQ keNKEd j7vNaf nID9nc"})
for article in articleList:
    print(article.div.article.h3.get_text())
    print(article.div.article.div.find_next_siblings('div')[0].div.a.get_text())
    time = article.div.article.div.find_next_siblings('div')[0].div.time
    if time:
        print(time.get_text())
    print('---------------------------------')
```
